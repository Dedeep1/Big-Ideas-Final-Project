# Below lies a map that gives the density of parks based on the area


```python
import pandas as pd
import matplotlib.pyplot as plt

link = "https://data.wprdc.org/datastore/dump/fa329e3d-89ff-4708-8ddf-81bfedcad11d"
df = pd.read_csv(link)


selected_columns = ['name', 'neighborhood', 'latitude', 'longitude', 'type', 'maintenance_responsibility']
df_selected = df[selected_columns]


df_selected = df_selected.dropna(subset=['latitude', 'longitude'])


neighborhood_counts = df_selected['neighborhood'].value_counts()


plt.figure(figsize=(20, 8))
plt.bar(neighborhood_counts.index, neighborhood_counts.values, color='blue')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Plots')
plt.title('Density of Plots for Each Neighborhood')
plt.xticks(rotation=90)  
plt.grid(True)
plt.show()
```
